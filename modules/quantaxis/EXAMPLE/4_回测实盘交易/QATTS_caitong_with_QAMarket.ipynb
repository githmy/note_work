```python
import QUANTAXIS as QA
import datetime

account = QA.QA_Account(
    user_cookie='tts',
    portfolio_cookie='tts',
    account_cookie='tts',
    broker=QA.BROKER_TYPE.TTS, 
    running_environment=QA.RUNNING_ENVIRONMENT.TTS
)

```

```python
market = QA.QA_Market(if_start_orderthreading=True)
market.connect(QA.BROKER_TYPE.TTS)
market.login(QA.BROKER_TYPE.TTS, account_cookie=account.account_cookie, account=account)
market.start()

```

```python
print(market.query_cash(account.account_cookie))

```

```python
print(market.query_assets(account.account_cookie))

```

```python
print(market.query_orders(account.account_cookie))
```

```python
print(market.query_order(account.account_cookie, '12384'))
```

```python
print(market.query_position(account.account_cookie))
```

```python
market.insert_order(account_cookie=account.account_cookie, code='600825', amount=100, amount_model=QA.AMOUNT_MODEL.BY_AMOUNT, order_model=QA.ORDER_MODEL.LIMIT,
                           frequence=QA.FREQUENCE.CURRENT, broker_name=account.broker, market_type=account.market_type,
                           towards=QA.ORDER_DIRECTION.BUY, price=1, time=datetime.datetime.now())
```
